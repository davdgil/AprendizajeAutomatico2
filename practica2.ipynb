## Ejercicio 1.1: Análisis de Código con Refinamiento

### Código a analizar

```python
def procesar_datos(datos):
    resultado = []
    for i in range(len(datos)):
        if datos[i] != None:
            if type(datos[i]) == str:
                resultado.append(datos[i].strip().lower())
            else:
                resultado.append(datos[i])
    return resultado

def buscar(lista, elemento):
    for i in range(len(lista)):
        if lista[i] == elemento:
            return i
    return -1
```

**Respuesta obtenida:**

> Aquí tienes un análisis de las dos funciones:
>
> **`procesar_datos(datos)`**
>
> Esta función recibe una lista llamada `datos` y devuelve una nueva lista `resultado` con los elementos procesados. Recorre la lista usando un índice numérico y realiza lo siguiente:
>
> - Si el elemento no es `None`, lo procesa.
> - Si el elemento es una cadena (`str`), le aplica `.strip()` para eliminar espacios en blanco al inicio y final, y `.lower()` para convertirlo a minúsculas.
> - Si no es un string, lo añade tal cual.
>
> En resumen, filtra los valores `None` y normaliza los strings.
>
> **Algunas mejoras posibles:**
> - Usar `is not None` en lugar de `!= None`, ya que es más idiomático en Python y evita problemas con objetos que sobrecargan `__eq__`.
> - Usar `isinstance(datos[i], str)` en lugar de `type(datos[i]) == str`, ya que `isinstance` también funciona con subclases de `str`.
> - En lugar de iterar con `range(len(datos))`, es más pythónico usar `for item in datos` directamente.
>
> **`buscar(lista, elemento)`**
>
> Implementa una búsqueda lineal. Recorre la lista y devuelve el índice del primer elemento que coincida con `elemento`. Si no lo encuentra, devuelve `-1`.
>
> Esto es equivalente a usar `lista.index(elemento)` con un try/except para manejar el caso en que el elemento no se encuentre. También podrías considerar usar el operador `in` para comprobar existencia antes de buscar.

En Python, la forma correcta es `is not None`, ya que `!=` invoca el método `__eq__` del objeto, lo que puede producir resultados inesperados con objetos que sobrecargan la comparación (por ejemplo, arrays de NumPy).
>   - **Sugerencia:** Cambiar `if datos[i] != None` por `if datos[i] is not None`.

> ### 2. Buenas prácticas / estilo
>
> - **Línea 2 — Severidad: Baja** — Se itera con `for i in range(len(datos))`, lo cual es un anti-pattern en Python. Lo pythónico es iterar directamente sobre los elementos.
>   - **Sugerencia:** Usar `for item in datos` en `procesar_datos` y `for i, item in enumerate(lista)` en `buscar`.
>
> - **Líneas 3-7 — Severidad: Baja** — Los `if` anidados pueden simplificarse. Se podría usar una list comprehension para `procesar_datos`.
>   - **Sugerencia:** Refactorizar con una list comprehension con condiciones.
>
> - **Línea 10 — Severidad: Baja** — Misma observación con `range(len(lista))` en `buscar`.
>   - **Sugerencia:** Usar `enumerate()`.
>
> ### 3. Optimización y rendimiento
>
> - **Líneas 10-13 — Severidad: Baja** — `buscar` reimplementa una búsqueda lineal que ya existe como built-in en Python (`list.index()`).
>   - **Sugerencia:** Usar `lista.index(elemento)` dentro de un bloque `try/except ValueError` para devolver `-1` si no se encuentra. Si se necesitan búsquedas frecuentes, considerar usar un `dict` o `set`.
>
> - **Complejidad:** `procesar_datos` es O(n) — está bien. `buscar` es O(n) — correcto para una lista no ordenada, pero podría mejorarse con estructuras de datos alternativas.
>
> ### 4. Seguridad y robustez
>
> - **Línea 1 — Severidad: Media** — No se valida que `datos` sea un iterable. Si se pasa un entero, por ejemplo, se produce un `TypeError` no controlado.
>   - **Sugerencia:** Añadir validación: `if not isinstance(datos, (list, tuple)): raise TypeError(...)`.
>
> - **Línea 9 — Severidad: Media** — Mismo problema con `lista` en `buscar`.
>   - **Sugerencia:** Validar el tipo de entrada.

## Ejercicio 1.3: Razonamiento con Chain of Thought

### Problema

> Una empresa de software tiene 3 equipos:
> - Equipo Frontend: 4 desarrolladores, cada uno puede completar 2 features/semana
> - Equipo Backend: 3 desarrolladores, cada uno puede completar 1.5 features/semana
> - Equipo QA: 2 testers, cada uno puede validar 5 features/semana
>
> Para el próximo release se necesitan 40 features desarrolladas y validadas.
> Considerando que QA solo puede validar features ya completadas:
> 1. ¿Cuántas semanas mínimo se necesitan?
> 2. ¿Hay algún cuello de botella? ¿Cuál?

---

### Resolución SIN Chain of Thought

**Prompt enviado a ChatGPT:**

> Una empresa de software tiene 3 equipos:
> - Frontend: 4 devs × 2 features/semana
> - Backend: 3 devs × 1.5 features/semana
> - QA: 2 testers × 5 features/semana
>
> Se necesitan 40 features desarrolladas y validadas. QA solo valida features ya completadas.
> ¿Cuántas semanas mínimo? ¿Hay cuello de botella?

**Respuesta obtenida:**

> **Capacidad de desarrollo:**
> - Frontend: 4 × 2 = 8 features/semana
> - Backend: 3 × 1.5 = 4.5 features/semana
> - Total: **12.5 features/semana**
>
> **Tiempo para desarrollar 40 features:** 40 ÷ 12.5 = 3.2 semanas → se redondea a **4 semanas**.
>
> **Capacidad de QA:** 2 × 5 = 10 features/semana. En 4 semanas QA puede validar 40 features.
>
> **Respuesta:** Se necesitan **4 semanas**. No hay un cuello de botella significativo, ya que QA tiene capacidad suficiente para validar todas las features en el tiempo disponible.

**Observaciones:** La respuesta es directa pero tiene un error: asume que QA puede validar en paralelo desde la semana 1, ignorando la restricción de que QA solo puede validar features **ya completadas**. No modela la dependencia temporal entre desarrollo y QA.

---

### Resolución CON Chain of Thought

**Prompt enviado a ChatGPT:**

> Resuelve el siguiente problema paso a paso. Muestra explícitamente cada paso de tu razonamiento antes de dar la respuesta final.
>
> Una empresa de software tiene 3 equipos:
> - Frontend: 4 devs × 2 features/semana = 8 features/semana
> - Backend: 3 devs × 1.5 features/semana = 4.5 features/semana
> - QA: 2 testers × 5 features/semana = 10 features/semana
>
> Se necesitan 40 features desarrolladas y validadas. QA solo puede validar features que ya han sido completadas por desarrollo.
>
> Paso 1: Calcula la capacidad de desarrollo semanal total.
> Paso 2: Calcula cuántas semanas necesita desarrollo para completar 40 features.
> Paso 3: Simula semana a semana cuántas features se desarrollan y cuántas puede validar QA.
> Paso 4: Determina la semana en que las 40 features están desarrolladas Y validadas.
> Paso 5: Identifica el cuello de botella.

**Respuesta obtenida:**

> **Paso 1: Capacidad de desarrollo semanal**
>
> - Frontend: 4 desarrolladores × 2 features/semana = **8 features/semana**
> - Backend: 3 desarrolladores × 1.5 features/semana = **4.5 features/semana**
> - **Capacidad total de desarrollo: 8 + 4.5 = 12.5 features/semana**
>
> **Paso 2: Semanas necesarias solo para desarrollo**
>
> - 40 features ÷ 12.5 features/semana = **3.2 semanas**
> - Como no se pueden tener semanas fraccionarias, desarrollo necesita **4 semanas** completas.
> - En la semana 4 solo se desarrollan 40 - 37.5 = **2.5 features** (el resto ya estaba hecho).
>
> **Paso 3: Simulación semana a semana**
>
> Capacidad de QA: 2 × 5 = **10 features/semana**. Pero QA solo puede validar features que ya están terminadas, así que en la semana 1 no hay nada que validar todavía (las features se completan al final de la semana).
>
> | Semana | Features desarrolladas esta semana | Total desarrolladas (acum.) | Features disponibles para QA | QA valida esta semana | Total validadas (acum.) |
> |--------|-----------------------------------|-----------------------------|------------------------------|-----------------------|-------------------------|
> | 1 | 12.5 | 12.5 | 0 (aún no hay features terminadas) | 0 | 0 |
> | 2 | 12.5 | 25.0 | 12.5 | 10 | 10 |
> | 3 | 12.5 | 37.5 | 15.0 (12.5 nuevas + 2.5 pendientes) | 10 | 20 |
> | 4 | 2.5 | 40.0 | 17.5 (12.5 nuevas + 5 pendientes) | 10 | 30 |
> | 5 | 0 | 40.0 | 10.0 (las que quedan) | 10 | 40 |
>
> **Paso 4: Resultado**
>
> - Al final de la **semana 4**, las 40 features están **desarrolladas**.
> - Al final de la **semana 5**, las 40 features están **validadas por QA**.
> - **Respuesta: se necesitan 5 semanas mínimo.**
>
> **Paso 5: Cuello de botella**
>
> El cuello de botella es **QA**, por dos razones:
>
> 1. **Dependencia secuencial:** QA no puede empezar hasta que desarrollo entregue features (pierde la semana 1 completa).
> 2. **Capacidad insuficiente:** Desarrollo produce 12.5 features/semana pero QA solo valida 10/semana. Esto crea un backlog de 2.5 features/semana que se acumula y obliga a QA a trabajar 1 semana extra después de que desarrollo haya terminado.
>
> **Para eliminar el cuello de botella** se necesitaría al menos 1 tester adicional (pasando de 10 a 15 features/semana de validación), o que QA pudiera empezar a validar durante la semana 1 en paralelo con desarrollo.

### Comparación de ambas respuestas

| Aspecto | SIN CoT | CON CoT |
|---------|---------|---------|
| **Respuesta final** | 4 semanas, sin cuello de botella | 5 semanas, QA es el cuello de botella |
| **Correctitud** | **Incorrecta** — ignora la dependencia secuencial | **Correcta** — modela la dependencia QA→Dev |
| **Razonamiento visible** | No, solo resultado directo | Sí, paso a paso con simulación tabular |
| **Identificación del cuello de botella** | "No hay" (error) | QA, con explicación cuantitativa |
| **Confianza en la respuesta** | Baja (sin forma de verificar) | Alta (se puede auditar cada paso) |

### Conclusión: ¿Cuándo usar CoT?

**Chain of Thought es esencial cuando:**
- El problema requiere **varios pasos de cálculo** interdependientes.
- Hay **dependencias temporales o lógicas** (en este caso, QA depende de desarrollo).
- Se necesita **auditabilidad**: poder verificar dónde falla el razonamiento.
- El resultado directo del modelo es **contraintuitivo o probablemente erróneo**.

**No es necesario cuando:**
- La pregunta tiene una respuesta factual directa (ej. "¿Cuál es la capital de Francia?").
- El cálculo es trivial y de un solo paso.

En este ejercicio, CoT corrigió un error de 1 semana (20% del resultado) y reveló un cuello de botella que el prompt directo ignoró por completo.

# Conclusiones

## Lecciones aprendidas

1. **El refinamiento iterativo no es opcional, es el método.** Un primer prompt casi nunca produce el resultado óptimo. El proceso de tres iteraciones (básico → estructurado → few-shot con restricciones) demostró mejoras cuantificables en cada paso. La clave está en analizar qué falta en cada respuesta y añadirlo al prompt siguiente.

2. **Los ejemplos (few-shot) son más poderosos que las instrucciones textuales.** En el Ejercicio 1.1, las instrucciones detalladas de la Iteración 2 produjeron buenas respuestas, pero la Iteración 3 con un ejemplo concreto logró adherencia perfecta al formato deseado. Los LLMs son imitadores de patrones más que seguidores de reglas.

3. **Chain of Thought corrige errores en problemas multi-paso.** Sin CoT, el modelo dio una respuesta incorrecta (4 semanas sin cuello de botella); con CoT, calculó correctamente 5 semanas e identificó QA como cuello de botella. La ganancia en problemas de razonamiento es sustancial.

4. **Las defensas de seguridad en system prompts funcionan, pero no son infalibles.** GPT-4o ignoró silenciosamente el prompt injection, mientras que Claude lo reconoció explícitamente, lo que paradójicamente revela información. La seguridad requiere las dos capas: instrucciones negativas Y tratamiento de comentarios/strings como código.

5. **Diferentes modelos tienen diferentes fortalezas.** No hay un "mejor modelo" universal; depende de los requisitos: GPT-4o para consistencia y seguridad, Claude para calidad de contenido y detalle.

## Técnica más útil para mí

La técnica que considero más valiosa es **el diseño de system prompts con estructura completa** (identidad + objetivo + formato + restricciones + seguridad + ejemplos). Es la técnica que más aplicabilidad tiene en el mundo real: permite construir asistentes reutilizables y profesionales, no solo obtener respuestas puntuales. Combinada con few-shot, produce resultados predecibles y de alta calidad.
