# Ejercicio 2: Experimentación con Embeddings y Similitud Semántica

Este notebook genera embeddings para un conjunto de frases y analiza las relaciones semánticas mediante similitud coseno.

## Paso 1: Configuración del entorno

Usamos Sentence-Transformers (opción local y gratuita)

```python
# Instalar dependencias si es necesario
# !pip install sentence-transformers numpy
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Cargar modelo multilingüe
model = SentenceTransformer('all-MiniLM-L6-v2')

def get_embedding(text):
    """Genera el embedding de un texto usando sentence-transformers."""
    return model.encode(text)

print(f"Modelo cargado: all-MiniLM-L6-v2")
print(f"Dimensión de embeddings: {model.get_sentence_embedding_dimension()}")
```

## Paso 2: Definir las frases de prueba

```python
frases = {
    # Grupo 1: Frases semánticamente similares (paráfrasis)
    "A1": "El gato se sentó en la alfombra",
    "A2": "Un felino descansaba sobre el tapete",
    "A3": "The cat sat on the mat",  # Misma idea, otro idioma

    # Grupo 2: Frases sobre tecnología
    "B1": "Python es un lenguaje de programación muy popular",
    "B2": "JavaScript se usa mucho para desarrollo web",
    "B3": "Los lenguajes de programación son herramientas esenciales",

    # Grupo 3: Negación y contraste
    "C1": "El restaurante tiene buena comida",
    "C2": "El restaurante no tiene buena comida",
    "C3": "La comida del restaurante es terrible",

    # Grupo 4: Frases sin relación
    "D1": "La fotosíntesis convierte luz solar en energía",
    "D2": "El precio del petróleo subió un 5% ayer",
}

# Mostrar las frases
for key, frase in frases.items():
    print(f"[{key}]: {frase}")
```

## Paso 3: Generar embeddings y calcular similitudes

```python
from numpy import dot
from numpy.linalg import norm

def cosine_similarity(a, b):
    """Calcula la similitud coseno entre dos vectores."""
    return dot(a, b) / (norm(a) * norm(b))

# Generar embeddings para todas las frases
embeddings = {}
for key, frase in frases.items():
    embeddings[key] = get_embedding(frase)
    print(f"[{key}] Embedding generado - dimensiones: {len(embeddings[key])}")
```

```python
# Calcular matriz de similitudes
print("\n--- MATRIZ DE SIMILITUD COSENO ---\n")
keys = list(frases.keys())
print(f"{'':>4}", end="")
for k in keys:
    print(f"{k:>8}", end="")
print()

for i, ki in enumerate(keys):
    print(f"{ki:>4}", end="")
    for j, kj in enumerate(keys):
        sim = cosine_similarity(embeddings[ki], embeddings[kj])
        print(f"{sim:>8.3f}", end="")
    print()
```

## Paso 4: Análisis de resultados específicos

```python
# Análisis de similitudes específicas
print("=" * 60)
print("ANÁLISIS DE SIMILITUDES ESPECÍFICAS")
print("=" * 60)

# Grupo A: Paráfrasis
print("\n--- GRUPO A: Paráfrasis ---")
sim_a1_a2 = cosine_similarity(embeddings["A1"], embeddings["A2"])
sim_a1_a3 = cosine_similarity(embeddings["A1"], embeddings["A3"])
sim_a2_a3 = cosine_similarity(embeddings["A2"], embeddings["A3"])
print(f"A1-A2 (paráfrasis español): {sim_a1_a2:.4f}")
print(f"A1-A3 (español-inglés):     {sim_a1_a3:.4f}")
print(f"A2-A3 (paráfrasis-inglés):  {sim_a2_a3:.4f}")

# Grupo B: Tecnología
print("\n--- GRUPO B: Tecnología ---")
sim_b1_b2 = cosine_similarity(embeddings["B1"], embeddings["B2"])
sim_b1_b3 = cosine_similarity(embeddings["B1"], embeddings["B3"])
sim_b2_b3 = cosine_similarity(embeddings["B2"], embeddings["B3"])
print(f"B1-B2 (Python-JavaScript):  {sim_b1_b2:.4f}")
print(f"B1-B3 (Python-general):     {sim_b1_b3:.4f}")
print(f"B2-B3 (JavaScript-general): {sim_b2_b3:.4f}")

# Grupo C: Negación
print("\n--- GRUPO C: Negación (IMPORTANTE) ---")
sim_c1_c2 = cosine_similarity(embeddings["C1"], embeddings["C2"])
sim_c1_c3 = cosine_similarity(embeddings["C1"], embeddings["C3"])
sim_c2_c3 = cosine_similarity(embeddings["C2"], embeddings["C3"])
print(f"C1-C2 (buena - NO buena):   {sim_c1_c2:.4f}  <- ¡Debería ser BAJA!")
print(f"C1-C3 (buena - terrible):   {sim_c1_c3:.4f}  <- Debería ser baja")
print(f"C2-C3 (no buena - terrible):{sim_c2_c3:.4f}  <- Deberían ser SIMILARES")

# Grupo D: Sin relación
print("\n--- GRUPO D: Sin relación ---")
sim_d1_d2 = cosine_similarity(embeddings["D1"], embeddings["D2"])
print(f"D1-D2 (fotosíntesis-petróleo): {sim_d1_d2:.4f} <- Debería ser MUY BAJA")

# Cross-grupo
print("\n--- CROSS-GRUPO (deberían ser bajas) ---")
sim_a1_b1 = cosine_similarity(embeddings["A1"], embeddings["B1"])
sim_a1_c1 = cosine_similarity(embeddings["A1"], embeddings["C1"])
sim_b1_d1 = cosine_similarity(embeddings["B1"], embeddings["D1"])
print(f"A1-B1 (gato-Python):       {sim_a1_b1:.4f}")
print(f"A1-C1 (gato-restaurante):  {sim_a1_c1:.4f}")
print(f"B1-D1 (Python-fotosíntesis):{sim_b1_d1:.4f}")
```

## Conclusiones

### Observaciones principales:

1. **Paráfrasis (A1-A2)**: Alta similitud (~0.75-0.85). Los modelos capturan bien que ambas frases expresan la misma idea.

2. **Cross-lingual (A1-A3)**: Similar a las paráfrasis en español. Los modelos multilingües capturan equivalencia semántica entre idiomas.

3. **Dominio temático (B1-B2-B3)**: Forman un cluster coherente con similitudes altas entre sí.

4. **PROBLEMA: Negación (C1-C2)**: Similitud ALTA pese a ser semánticamente opuestas. Los embeddings NO capturan bien la negación. Esta es una limitación conocida.

5. **Frases sin relación (D1-D2)**: Baja similitud como esperado.

### Limitación importante:
Los embeddings representan el "tema" o "contenido léxico" de una frase, pero NO capturan bien la polaridad, negación o contradicción. Esto es importante para diseñar sistemas RAG: no podemos confiar solo en similitud de embeddings para detectar contradicciones.
